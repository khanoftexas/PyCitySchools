# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing reading performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
school_data.head()
```

```python
# Read Student Data File and store into Pandas Data Frames
student_data = pd.read_csv(student_data_to_load)
student_data.head()
```

```python
# Combine the data into a single dataset
#school_data.rename(columns = {'name': 'school_name'}, inplace = True)
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name","school_name"])
school_data_complete
```

```python
# Calculate the total number of schools
dist_total_schools = school_data["school_name"].nunique()
dist_total_schools
```

```python
# Calculate the total number of students
dist_total_students = school_data["size"].sum()
dist_total_students
```

```python
# Calculate the total budget
dist_total_budget = school_data["budget"].sum()
dist_total_budget 
```

```python
# Calculate the average math score
dist_average_math_score = round(student_data["math_score"].mean(),6)
dist_average_math_score
```

```python
# Calculate the average reading score
dist_average_reading_score = round(student_data["reading_score"].mean(),6)
dist_average_reading_score
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
dist_passing_math_pct = round((student_data.loc[student_data["math_score"] >= 70, "math_score"].count())/(school_data["size"].sum())*100,6)
dist_passing_math_pct
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
dist_passing_reading_pct = round((student_data.loc[student_data["reading_score"] >= 70, "reading_score"].count())/(school_data["size"].sum())*100,6)
dist_passing_reading_pct
```

```python
# Calculate the overall passing rate (overall average score), i.e. (avg. reading score + avg. reading score)/2
dist_overall_passing_rate =round((dist_average_math_score + dist_average_reading_score)/2,6)
dist_overall_passing_rate
```

```python
# Create a dataframe to hold the above results
district_summary = pd.DataFrame({   
    "Total Schools": [dist_total_schools],
    "Total Students": [dist_total_students],
    "Total Budget": [dist_total_budget],
    "Average Math Score": [dist_average_math_score],
    "Average Reading Score": [dist_average_reading_score],
    "% Passing Math": [dist_passing_math_pct],
    "% Passing Reading":[dist_passing_reading_pct], 
    "Overall Passing Rate": [dist_overall_passing_rate]
})
```

```python
 # Optional: give the displayed data cleaner formatting
district_summary_formatted = district_summary.style.format({
    "Total Students": "{:,}".format,
    "Total Budget": "${:,.2f}".format, 
    "Average Math Score": "{:.6f}".format, 
    "Average reading Score": "{:.6f}".format,
    "% Passing reading": "{:.6f}%".format,
    "% Passing Reading": "{:.6f}%".format,
    "Overall Passing Rate": "{:.6f}".format})

district_summary_formatted
```

```python
# School Name
school_name_group = school_data_complete.set_index("school_name")
school_name_group.head()
```

```python
# School Type
school_types = school_data.set_index("school_name")["type"]
school_types
```

```python
# Total Students
school_total_students = school_data_complete["school_name"].value_counts()
school_total_students
```

```python
# Total School Budget
school_total_budget = school_data.groupby(["school_name"]).sum()["budget"]
school_total_budget
```

```python
# Per Student Budget
school_per_student_budget = school_total_budget / school_total_students
school_per_student_budget
```

```python
# Average math Score
school_average_mathc_score = school_data_complete.groupby(["school_name"]).mean()["math_score"]
school_average_math_score
```

```python
# Average Reading Score
school_average_reading_score = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
school_average_reading_score
```

```python
# % Passing Math
school_passing_reading = school_data_complete[(school_data_complete["math_score"] >= 70)]
school_passing_reading_pct = round((school_passing_reading.groupby(["school_name"]).count()["student_name"])/(school_data_complete["school_name"].value_counts())*100,6)
school_passing_reading_pct
```

```python
# % Passing Reading
school_passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)]
school_passing_reading_pct = round((school_passing_reading.groupby(["school_name"]).count()["student_name"])/(school_data_complete["school_name"].value_counts())*100,6)
school_passing_reading_pct
```

```python
# Overall Passing Rate (Average of the above two)
school_overall_passing = (school_passing_math_pct+school_passing_reading_pct)/2
school_overall_passing
```

```python
# Create a dataframe to hold the above results
school_summary = pd.DataFrame({"School Type": school_types,
                                "Total Students": school_total_students,
                                "Total School Budget": school_total_budget,
                                "Per Student Budget": school_per_student_budget,
                                "Average Math Score": school_average_math_score,
                                "Average Reading Score": school_average_reading_score,
                                "% Passing Math": school_passing_reading_pct,
                                "% Passing Reading": school_passing_reading_pct,
                                "Overall Passing Rate": school_overall_passing})
school_summary_formatted = school_summary.style.format({
    "Total Students": "{:,}".format,
    "Total School Budget": "${:,.2f}".format, 
    "Per Student Budget": "${:,.2f}".format,
    "% Passing Math": "{:.6f}%".format,
    "% Passing Reading": "{:.6f}%".format
})

school_summary_formatted
```
